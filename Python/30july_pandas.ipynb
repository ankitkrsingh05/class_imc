```python
csv_file_path = '/Users/ankitkumarsingh/Documents/imarticus/class_imc/Python/employee_details.csv'
df = pd.read_csv(csv_file_path)
```

```python
df.columns = [cols.lower().replace('\t','') for cols in df.columns]
```

```python
# write a program to find the max salary in the department id 90 or 70
# write a program to find the name of the employee in department id 
# 60 or 90 or 70 earning the maximum
print ([(df['department_id'] == 90) | (df['department_id'] == 70)]) # this is the condition 
# here are 2 ways to solve the problem
df[(df['department_id'] == 90) | (df['department_id'] == 70)]['salary'].max() # over here we are doing dataframe slicing then selecting the column and then applying max
print (df['salary'][(df['department_id'] == 90) | (df['department_id'] == 70)].max()) # over here we have got the series then we are slicing the series and then applying max on the selected series values
# Next problem is to find the person with the maximum salary
# soo we will use the max salary value , compare it to all the salary,
# we get a list with only 1 true value (concidering the max value is unique)
# we use the list to get the row with the maximum salary
df[['first_name','last_name']][df['salary'] == df['salary'][(df['department_id'] == 90) | (df['department_id'] == 70)].max()]

# df['salary'][(df['department_id'] == 90) | (df['department_id'] == 70)].max() # = 63090
# [df['salary'] == 63090]
```

```python
# write a program to find all the employees 
# whose salary is greater than the avg salary of departmentid 70

# steps:
    # first try to select only the salary and department id
    # try to add the condition of department id 70
    # find the avg salary of department id 70
    # use this single value of avg salary 
        #to find the rows whose salary > avg salary of department 70

[df['department_id'] == 70] # this is the condition to get the department id 70 rows
df['salary'][df['department_id'] == 70] # this is to use the above condition to get the salary values (series)
mean_sal_dep_70 = df['salary'][df['department_id'] == 70].mean() # this is to find the average of the values generated in above code
print(mean_sal_dep_70) # storing the values in a variable

df[df['salary'] > mean_sal_dep_70] # 
df[df['salary'] > df['salary'][df['department_id'] == 70].mean()] # consolidated code to do the same in a single line

```

```python
# pandas loc and iloc
    # loc : [<row selection,column selection>],selection based on condition examples, range of rows,updating the values,
    # iloc: rows and columns selection [index selection,column selection]
# difference is that iloc is index based selection and loc is label based selection
        # the end value in iloc is exclusive
        # the end value in loc is inclusive
    # in loc we can apply condition based filtering whereas in iloc we cannot
```

```python
# [<row selection,column selection>]
print (df.loc[1:5,['first_name','last_name']])
print (df.iloc[0:5,1:3])
```

```python
# iloc
df.iloc[[2,4,6,7]] # one way to select a list of rows
print (df.iloc[0:5,1:3]) # based on rows and columns indexes

# loc
# condition 
df.loc[df['department_id'] == 70]
# update the values
df['salary'].loc[df['department_id'] == 70] = df['salary'] + 10
```
