```python
P(getting an even on A) = {(2, 1), (2, 2), (2, 3), (2, 4), (2, 5), (2, 6),
                         (4, 1), (4, 2), (4, 3), (4, 4), (4, 5), (4, 6),
                         (6, 1), (6, 2), (6, 3), (6, 4), (6, 5), (6, 6)}
```

```python
P(getting an even on B) = { (1, 2), (1, 4), (1, 6),
 (2, 2), (2, 4),  (2, 6),
(3, 2),(3, 4),  (3, 6),
(4, 2), (4, 4),  (4, 6),
(5, 2), (5, 4),  (5, 6),
(6, 2), (6, 4),  (6, 6)}
```

```python
P(getting a even on both the dices) = {
        (2, 2),(2, 4),(2, 6),
        (4, 2), (4, 4),  (4, 6),
        (6, 2), (6, 4),  (6, 6)
}
```

```python
# Intersection
P(getting an even on both dices) = P(getting an even on dice A) * P(getting an even on dice B)
```

```python
# Union
P(getting an even on A or getting an even on B or getting an even on both) = P(getting an even on dice A) + P(getting an even on dice B)
```

```python
# trial
def trial_diceroll():
    outcome = random.randint(1,6)
    # print (f'for the trail of dice roll the outcome is {outcome}')
    return outcome
```

```python
print (trial_diceroll())
```

```python
# event of rolling dice
event_outcomes = []
for i in range(10000):
    event_outcomes.append(trial_diceroll())
```
