Shapiro-Wilk Test: Tests whether data is normally distributed. Null hypothesis (H0): Data is normally distributed.

Anderson-Darling Test: Provides a test statistic to determine normality. The critical values at various significance levels are provided. If the test statistic exceeds the critical value for a given significance level, reject the null hypothesis (data is not normal).

Kolmogorov-Smirnov Test: Compares the data to a normal distribution. Null hypothesis (H0): Data is drawn from the given distribution.

D'Agostino and Pearson's Test: Combines skewness and kurtosis to produce an omnibus test for normality. Null hypothesis (H0): Data is normally distributed.

Q-Q Plot: Visual method to assess normality. If data points fall on the 45-degree line, the data is normally distributed.

Histogram with KDE (Kernel Density Estimate): Provides a visual representation of the distribution of data. A normal distribution should show a bell-shaped curve.

One-Sample T-Test (t-test for a single mean)
Formula: t = (x̄ - μ) / (s / √n)

where: t = t-statistic x̄ = sample mean μ = population mean (known) s = sample standard deviation n = sample size

Independent Samples T-Test (t-test for two independent means)
Formula: t = (x̄1 - x̄2) / (s_p √(1/n1 + 1/n2))

where: t = t-statistic x̄1 and x̄2 = sample means for each group s_p = pooled sample standard deviation n1 and n2 = sample sizes for each group

Paired Samples T-Test (t-test for two related means)
Formula: t = (∑(d_i) - 0) / (s_d √(n-1))

where: t = t-statistic d_i = differences between paired observations s_d = standard deviation of differences n = number of pairs

Equal Variance T-Test
Formula: t = (x̄1 - x̄2) / √((s1^2 + s2^2) / n1 + n2 - 2)

where: t = t-statistic x̄1 and x̄2 = sample means for each group s1 and s2 = sample standard deviations for each group n1 and n2 = sample sizes for each group

Unequal Variance T-Test (Welch’s t-test)
Formula: t = (x̄1 - x̄2) / √((s1^2 / n1 + s2^2 / n2)

where: t = t-statistic x̄1 and x̄2 = sample means for each group s1 and s2 = sample standard deviations for each group n1 and n2 = sample sizes for each group

Note: These formulas assume normality and equal or unequal variances, depending on the type of t-test. Additionally, the degrees of freedom (df) may vary depending on the type of t-test and the assumptions met.


```python
import numpy as np
import scipy.stats as stats
import pandas as pd

# One-Sample t-test
# Explanation: This test compares the mean of a single group to a known value or population mean.
print("=== One-Sample t-test ===")
# Example: We have a sample with a known population mean of 50
data_one_sample = np.array([52, 48, 49, 51, 50, 53, 47, 50, 49])
population_mean = 50

t_stat, p_value = stats.ttest_1samp(data_one_sample, population_mean)
print(f"Data: {data_one_sample}")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
print("Conclusion: Reject the null hypothesis if P-value < 0.05")
print()

# Independent Two-Sample t-test
# Explanation: This test compares the means of two independent groups to see if they are significantly different.
print("=== Independent Two-Sample t-test ===")
# Example: We have two independent groups with their own sample data
group1 = np.array([20, 22, 19, 24, 21, 23, 25, 22])
group2 = np.array([30, 32, 29, 31, 28, 33, 30, 31])

t_stat, p_value = stats.ttest_ind(group1, group2)
print(f"Group 1: {group1}")
print(f"Group 2: {group2}")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
print("Conclusion: Reject the null hypothesis if P-value < 0.05")
print()

# Paired Sample t-test
# Explanation: This test compares the means of two related groups to determine if there is a significant difference.
print("=== Paired Sample t-test ===")
# Example: We have the same subjects measured before and after treatment
before_treatment = np.array([85, 88, 89, 92, 87, 85, 90, 91])
after_treatment = np.array([88, 90, 91, 94, 89, 87, 93, 92])

t_stat, p_value = stats.ttest_rel(before_treatment, after_treatment)
print(f"Before Treatment: {before_treatment}")
print(f"After Treatment: {after_treatment}")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")
print("Conclusion: Reject the null hypothesis if P-value < 0.05")
print()

# Summary
summary = pd.DataFrame({
    'Test Type': ['One-Sample t-test', 'Independent Two-Sample t-test', 'Paired Sample t-test'],
    'T-Statistic': [stats.ttest_1samp(data_one_sample, population_mean)[0], 
                    stats.ttest_ind(group1, group2)[0], 
                    stats.ttest_rel(before_treatment, after_treatment)[0]],
    'P-Value': [stats.ttest_1samp(data_one_sample, population_mean)[1], 
                stats.ttest_ind(group1, group2)[1], 
                stats.ttest_rel(before_treatment, after_treatment)[1]],
})

print("=== Summary ===")
print(summary)
```

```python
Chi square formula
The chi-square formula is used to calculate the chi-square statistic (X²) for testing the independence of two categorical variables. The formula is:

X² = ∑ (Oi - Ei)² / Ei

Where:

Oi is the observed frequency (count) for each cell in the contingency table
Ei is the expected frequency (count) for each cell under the null hypothesis of independence
The summation (∑) is taken over all cells in the contingency table
This formula calculates the squared difference between the observed and expected frequencies for each cell, and then divides the result by the expected frequency. The sum of these values is the chi-square statistic.

Note that the expected frequencies (Ei) are typically calculated based on the marginal totals of the contingency table, assuming independence between the variables. The chi-square statistic is then compared to a critical value from the chi-square distribution to determine the significance of the association between the variables.
```

Anova formula
The ANOVA (Analysis of Variance) formula is used to compare the means of three or more groups to determine if there are significant differences between them. The formula is based on the following components:

Source of Variation:
Between Groups (SSB): Sum of Squares between groups, representing the variation due to differences between group means.
Within Groups (SSE): Sum of Squares within groups, representing the variation due to individual data points within each group.
Total (SST): Total Sum of Squares, representing the total variation in the data.
Degrees of Freedom:
df1 (k-1): Degrees of freedom for the between-groups variation, where k is the number of groups.
df2 (N-k): Degrees of freedom for the within-groups variation, where N is the total sample size.
Mean Squares:
MSB (SSB / df1): Mean Square between groups, representing the average squared difference between group means.
MSE (SSE / df2): Mean Square within groups, representing the average squared difference within each group.
F-Statistic:
F = MSB / MSE: The F-statistic, which is used to determine the significance of the differences between group means.
The ANOVA formula is:

SSB = Σnj² - (Σnj)² / N

SSE = Σ(xi - xī)²

SST = SSB + SSE

MSB = SSB / (k-1)

MSE = SSE / (N-k)

F = MSB / MSE

where:

nj is the number of observations in each group
xi is an individual data point
xī is the mean of each group
N is the total sample size
k is the number of groups
Note that the ANOVA formula assumes normality of the data and equal variances across groups. If these assumptions are not met, alternative tests or transformations may be necessary.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.graphics.gofplots import qqplot

# Generate random data
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# 1. Shapiro-Wilk Test
shapiro_test = stats.shapiro(data)
print(f'Shapiro-Wilk Test: Statistic={shapiro_test.statistic}, p-value={shapiro_test.pvalue}')

# 2. Anderson-Darling Test
anderson_test = stats.anderson(data, dist='norm')
print(f'Anderson-Darling Test: Statistic={anderson_test.statistic}, Critical Values={anderson_test.critical_values}, Significance Levels={anderson_test.significance_level}')

# 3. Kolmogorov-Smirnov Test
ks_test = stats.kstest(data, 'norm')
print(f'Kolmogorov-Smirnov Test: Statistic={ks_test.statistic}, p-value={ks_test.pvalue}')

# 4. D'Agostino and Pearson's Test
dagostino_test = stats.normaltest(data)
print(f'D\'Agostino and Pearson\'s Test: Statistic={dagostino_test.statistic}, p-value={dagostino_test.pvalue}')

# 5. Q-Q Plot
qqplot(data, line='s')
plt.title("Q-Q Plot")
plt.show()

# 6. Histogram
sns.histplot(data, kde=True)
plt.title("Histogram with KDE")
plt.show()
```

One-sample t-test:

Hypothesis: Tests whether the average age of all Titanic passengers is significantly different from a population mean (e.g., 30 years).
The ttest_1samp function is used to compare the sample mean to a known value.
Independent t-test (Two-sample):

Hypothesis: Tests whether there is a significant difference between the average ages of male and female passengers.
The ttest_ind function is used for this test. equal_var=False indicates the use of Welch's t-test, which does not assume equal variances.
Paired t-test:

Hypothesis: Tests if there's a significant difference in the ages of passengers between Class 1 and Class 2 (artificial example).
The ttest_rel function is used to compare two related groups. Here, we've paired the data artificially by trimming the groups to the same size.

```python
# t- test
    # 

import pandas as pd
import seaborn as sns
from scipy import stats

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Drop rows where 'age' or 'sex' is NaN
titanic = titanic.dropna(subset=['age', 'sex'])

# Separate the data into two groups: males and females
age_male = titanic[titanic['sex'] == 'male']['age']
age_female = titanic[titanic['sex'] == 'female']['age']

### 1. One-sample t-test ###
# Hypothesis: Is the average age of passengers significantly different from 30 years?

# Population mean to test against
population_mean = 30

# Perform the One-sample t-test
t_stat_one_sample, p_value_one_sample = stats.ttest_1samp(titanic['age'], population_mean)

print("One-sample t-test:")
print(f"t-statistic: {t_stat_one_sample}")
print(f"p-value: {p_value_one_sample}\n")

# Interpretation
if p_value_one_sample < 0.05:
    print("The average age of passengers is significantly different from 30 years.\n")
else:
    print("The average age of passengers is not significantly different from 30 years.\n")


### 2. Independent t-test (Two-sample) ###
# Hypothesis: Is there a significant difference between the average ages of male and female passengers?

# Perform the Independent t-test (Welch's t-test)
t_stat_ind, p_value_ind = stats.ttest_ind(age_male, age_female, equal_var=False)

print("Independent t-test:")
print(f"t-statistic: {t_stat_ind}")
print(f"p-value: {p_value_ind}\n")

# Interpretation
if p_value_ind < 0.05:
    print("There is a significant difference between the average age of male and female passengers.\n")
else:
    print("There is no significant difference between the average age of male and female passengers.\n")


### 3. Paired t-test ###
# Hypothesis: Is there a significant difference between the age of passengers in different classes?

# Since we don't have before-and-after data, we'll create an artificial scenario:
# Let's test if there's a significant difference between the age of passengers in Class 1 and Class 2.
# We need paired samples, so we'll only consider passengers who are present in both Class 1 and Class 2.

# Create data for the paired test by selecting passengers present in both classes
class_1 = titanic[titanic['class'] == 'First'][['age']].reset_index(drop=True)
class_2 = titanic[titanic['class'] == 'Second'][['age']].reset_index(drop=True)

# Ensure the paired samples have the same size by trimming the larger dataset
min_size = min(len(class_1), len(class_2))
class_1 = class_1.iloc[:min_size]
class_2 = class_2.iloc[:min_size]

# Perform the Paired t-test
t_stat_paired, p_value_paired = stats.ttest_rel(class_1['age'], class_2['age'])

print("Paired t-test:")
print(f"t-statistic: {t_stat_paired}")
print(f"p-value: {p_value_paired}\n")

# Interpretation
if p_value_paired < 0.05:
    print("There is a significant difference in the ages of passengers between Class 1 and Class 2.\n")
else:
    print("There is no significant difference in the ages of passengers between Class 1 and Class 2.\n")
```

```python
# goodness of fit
import numpy as np
from scipy import stats

# Observed frequencies from 60 rolls of a die
observed_frequencies = np.array([8, 9, 10, 11, 12, 10])

# Expected frequencies for a fair die (1/6th of the total rolls)
total_rolls = observed_frequencies.sum()
expected_frequencies = np.full(6, total_rolls / 6)

print("Observed Frequencies:", observed_frequencies)
print("Expected Frequencies:", expected_frequencies)

# Perform the Chi-Square Goodness-of-Fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

print("\nChi-Square Goodness-of-Fit Test Results:")
print(f"Chi2 Statistic: {chi2_stat}")
print(f"p-value: {p_value}")

# Interpretation
alpha = 0.05  # significance level

if p_value < alpha:
    print("The die is not fair. (Reject H₀)")
else:
    print("The die is fair. (Fail to reject H₀)")
```

```python
import pandas as pd
import seaborn as sns
from scipy import stats

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Count the observed number of passengers in each class
observed_frequencies = titanic['class'].value_counts().sort_index()

# Define the expected distribution: 30% First, 30% Second, 40% Third
total_passengers = observed_frequencies.sum()
expected_frequencies = total_passengers * pd.Series([0.3, 0.3, 0.4], index=observed_frequencies.index)

print("Observed Frequencies:")
print(observed_frequencies)
print("\nExpected Frequencies:")
print(expected_frequencies)

# Perform the Chi-Square Goodness-of-Fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

print("\nChi-Square Goodness-of-Fit Test Results:")
print(f"Chi2 Statistic: {chi2_stat}")
print(f"p-value: {p_value}")

# Interpretation
alpha = 0.05  # significance level

if p_value < alpha:
    print("The distribution of passengers across classes significantly differs from the expected distribution. (Reject H₀)")
else:
    print("The distribution of passengers across classes does not significantly differ from the expected distribution. (Fail to reject H₀)")
```

1. ANOVA (Analysis of Variance)
Overview: ANOVA is a statistical test used to determine whether there are any statistically significant differences between the means of three or more independent (unrelated) groups. It’s an extension of the t-test when you have more than two groups.

When to Use:

When comparing the means of three or more independent groups.
Example Scenario on Titanic Dataset: We’ll use ANOVA to test whether the average age of passengers differs significantly across different classes (First, Second, Third).


ANOVA Hypotheses:
Null Hypothesis (H₀): The means of the groups (First, Second, and Third class) are equal.
Alternative Hypothesis (H₁): At least one group mean is different.
F-statistic: Measures the ratio of variance between the group means to the variance within the groups. A higher F-statistic indicates a greater degree of separation between the means.
p-value: If the p-value is less than 0.05, we reject the null hypothesis, indicating that there is a significant difference in means.

```python
# anova

import pandas as pd
import seaborn as sns
from scipy import stats

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Drop rows where 'age' or 'class' is NaN
titanic = titanic.dropna(subset=['age', 'class'])

# Separate the data by passenger class
age_first = titanic[titanic['class'] == 'First']['age']
age_second = titanic[titanic['class'] == 'Second']['age']
age_third = titanic[titanic['class'] == 'Third']['age']

# Perform the ANOVA
f_stat, p_value = stats.f_oneway(age_first, age_second, age_third)

print("ANOVA Test:")
print(f"F-statistic: {f_stat}")
print(f"p-value: {p_value}\n")

# Interpretation
if p_value < 0.05:
    print("There is a significant difference in the average age across passenger classes.")
else:
    print("There is no significant difference in the average age across passenger classes.")
```
